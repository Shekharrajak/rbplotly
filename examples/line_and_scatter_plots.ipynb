```ruby
require 'rbplotly'

n = 100

x  = (0..n).map { |i| i.to_f / n }
y0 = (0..n).map { rand(-2.0..2.0) + 5 }
y1 = (0..n).map { rand(-2.0..2.0) }

trace0 = { x: x, y: y0, type: 'scatter', mode: 'markers' }
trace1 = trace0.merge(y: y1, mode: 'markers+lines')

plot = Plotly::Plot.new(data: [trace0, trace1])
plot.show
```

```ruby
plot.data.last.y = (0..n).map { rand(-0.5..0.5) }
plot.show
```

```ruby
new_trace = { x: x, y: (0..n).map { rand(-2.0..2.0) - 5 }, type: 'scatter', mode: 'lines' }
plot.add_data(new_trace)
plot.show
```

```ruby
plot.layout.xaxis = { title: 'x title' }
plot.layout.yaxis = { title: 'y title' }

plot.show
```
